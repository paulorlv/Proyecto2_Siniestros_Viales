Ahora vamos a extraer datos de una API del sitio web del Gobierno de Buenos Aires y formular una función para obtener los barrios en base a las coordenadas dadas por las columnas 'Pos X' y 'Pos Y'

```python
def get_vecindario(fila):
    """
    Recupera el nombre del vecindario según las coordenadas geográficas mediante una API externa.

    Parametros:
    -----------
       fila (pandas.Series): una fila de un DataFrame que contiene las columnas 'Pos X' y 'Pos Y'.

    Retorna:
    --------
        str: El nombre del vecindario si se encuentra, o "Desconocido" si no se encuentra o hay un error.
    """
    # URL base para la API externa
    base_url = "https://datosabiertos-usig-apis.buenosaires.gob.ar/datos_utiles"
    
    # Parámetros de consulta para la solicitud de API
    parametros = {
        "x": fila["Pos X"],
        "y": fila["Pos Y"],
        "formato": "json"
    }
    
    # Creamos la URL completa con parámetros de consulta
    full_url = urljoin(base_url, '?' + urlencode(parametros))

    try:
        # Realizamos la solicitud HTTP
        response = requests.get(full_url)
        
        # Verificamos el código de estado de la respuesta.
        response.raise_for_status()
        
        # Analiza JSON y obtiene el nombre del vecindario si está presente
        data = response.json()
        return data.get("barrio", "Desconocido")
    
    except requests.exceptions.RequestException as e:
        print("Error al realizar la solicitud:", e)
        return "Desconocido"
```

```python
# Aplicamos la funcion en la nueva columna'Barrio'
accidentes["Barrio"] = accidentes[["Pos Y", "Pos X"]].apply(get_vecindario, axis=1)
# Observamos la columna
accidentes["Barrio"]
```

Web Scrapping

Ahora. Vamos a hacer webcrapping para sacar información de la población de la Ciudad de Buenos Aires

```python
# Obtenemos la URL con la información
url = 'https://es.wikipedia.org/wiki/Buenos_Aires'
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

Buscamos la tabla en la URL que contiene la información que necesitamos.

```python
# Una variable llamada tabla se inicializa como None
tabla = None

# Buscamos todas las etiquetas de tabla (<table>) en el contenido HTML
tablas = soup.find_all('table')

# Itera a través de todas las tablas que se encuentran en la página web.
for table in tablas:
    
# Busca una etiqueta de encabezado (<th>) dentro de la tabla que tiene los atributos colspan='3' y class='navbox-title'
    encabezado = table.find('th', {'colspan': '3', 'class': 'navbox-title'})
    
    # Comprueba si se encontró un encabezado y si el texto del encabezado contiene "Población histórica"
    if encabezado and "Población histórica" in encabezado.text:
        # Si se cumple la condición anterior, asigne la tabla actual a la variable tabla y salga del bucle for con break
        tabla = table
        break
```

Cargamos la tabla

```python
tabla
```

Guardamos la tabla en un df

```python
# Creamos listas para almacenar las columnas de la tabla.
años = []
poblaciones = []
cambios = []

# Inicializamos una variable en None
año_cualquiera = None

# Iteramos a través de las filas de la tabla.
for row in tabla.find_all('tr'):
    # Buscamos tanto th (encabezados) como td (celdas de datos)
    columns = row.find_all(['th', 'td'])

    #  Asegura de que haya suficientes columnas para extraer el año, la población y el cambio porcentual
    if len(columns) >= 3:
        año_cualquiera = columns[0].get_text(strip=True)  # Get Year
        poblacion = columns[1].get_text(strip=True)  # Get Population
        cambio = columns[2].get_text(strip=True)  # Get population changes

        # Agregamos valores a las listas respectivas
        años.append(año_cualquiera)  
        poblaciones.append(poblacion)  
        cambios.append(cambio)  

# Se crea un DataFrame de Pandas con los datos extraídos
Censo = pd.DataFrame({'Año': años, 'Poblacion': poblaciones, 'Cambio': cambios})

print("Dataframe guardado correctamente")
```

```python
Censo
```

Corregimos los datos del Dataframe.

```python
# Se elimina la primera fila.
Censo = Censo.iloc[1:]

# Se elimina la columna 'Cambio'
Censo = Censo.drop('Cambio', axis=1)

# Los espacios de separación de miles se eliminan en la columna 'Población'
Censo['Poblacion'] = Censo['Poblacion'].str.replace(' ', '').str.replace('\xa0', '').astype(int)

# Convertimos la columna 'Año' a tipo entero
Censo['Año'] = Censo['Año'].astype(int)

# Reseteamos el indice
Censo = Censo.reset_index(drop=True)
```

Como solo tenemos datos de los censos de 2010 y 2022, utilizaremos el crecimiento poblacional anual promedio para estimar la población año por año.

```python
# Obtenemos la población para el año 2010 si hay datos.
pop_2010 = Censo.loc[Censo['Año'] == 2010, 'Poblacion'].values[0] if len(Censo.loc[Censo['Año'] == 2010]) > 0 else None

# Obtenemos la población para el año 2022 si hay datos.
pop_2022 = Censo.loc[Censo['Año'] == 2022, 'Poblacion'].values[0] if len(Censo.loc[Censo['Año'] == 2022]) > 0 else None

# Verificamos si se obtuvieron datos antes de continuar
if pop_2010 is not None and pop_2022 is not None:
    # Calculamos la tasa de crecimiento anual promedio
    tasa_crecimiento = (pop_2022 / pop_2010) ** (1 / (2022 - 2010)) - 1

    # Generamos la lista de años del 2010 al 2022
    años = list(range(2010, 2023))

    # Calculamos la población para cada año usando la tasa de crecimiento.
    pobla_personas_año = [int(pop_2010 * (1 + tasa_crecimiento) ** (year - 2010)) for year in años]

    # Mostramos lista
    print(pobla_personas_año)
else:
    print("No hay datos para los valores de 2010 y 2022.")
```

```python
data = {'Año': años, 'Población':pobla_personas_año}
Pobla_BsAs = pd.DataFrame(data)
Pobla_BsAs
```
