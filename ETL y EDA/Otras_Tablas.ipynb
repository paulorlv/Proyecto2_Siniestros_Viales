```python
df = pd.read_csv("data/csv/accidentes_limpio.csv")
```

```python
df['Numero_Comuna'] = df['Comuna'].str.extract('(\d+)').astype(int)

# Creamos un nuevo DataFrame 'commune' con las columnas 'Comuna' y 'Comuna_Id'
commune = df[['Comuna', 'Numero_Comuna']].drop_duplicates().reset_index(drop=True)
commune.columns = ['Comuna', 'Comuna_Id']
order = ['Comuna_Id', 'Comuna']
commune = commune[order]

commune = commune.sort_values(by='Comuna_Id').reset_index(drop=True)
commune
```

```python
df.head(3)
```

```python
commune_unique = df['Numero_Comuna'].unique()

# Creamos un diccionario para almacenar identificadores de barrio por comuna
ndh_id_dict = {}

# Iteramos sobre cada comuna única
for Numero_Comuna in commune_unique:
    #Filtramos el DataFrame para la comuna actual
    df_commune = df[df['Numero_Comuna'] == Numero_Comuna]
    
    # Obtenemos los barrios únicos de la comuna.
    nbh_unique = df_commune['Barrio'].unique()
    
    # Asignamos un identificador único a cada nbh dentro de la comuna
    for i, nbh in enumerate(nbh_unique, start=1):
        barrio_id = f'{Numero_Comuna}-{i}'
        ndh_id_dict[nbh] = barrio_id

# Creamos un DataFrame para la tabla 
nbh = pd.DataFrame({'Barrio_Id': list(ndh_id_dict.values()), 'Barrio': list(ndh_id_dict.keys())})

nbh = nbh.sort_values(by='Barrio_Id').reset_index(drop=True)

# Visualizamos
nbh
```

```python
season = df[["Estación"]].drop_duplicates().reset_index(drop=True)
season['Estación_Id'] = season.index + 1
order = ['Estación_Id', 'Estación']
season = season[order]
season
```

```python
age_range = df[['Rango Etario']].drop_duplicates()
age_range = age_range.sort_values(by='Rango Etario').reset_index(drop=True)
age_range['Rango_Etario_Id'] = age_range.index + 1
order = ['Rango_Etario_Id', 'Rango Etario']
age_range = age_range[order]
age_range
```

```python
df['Participantes'].unique()
```

```python
participants = df[["Participantes"]].drop_duplicates().reset_index(drop=True)
participants['Participantes_Id'] = participants.index + 1
participants.loc[participants['Participantes'] == 'SD-SD', 'Participantes_Id'] = 0
participants = participants.sort_values(by='Participantes_Id').reset_index(drop=True)
order = ['Participantes_Id', 'Participantes']
participants = participants[order]
participants
```

```python
rol = df[["Rol"]].drop_duplicates().reset_index(drop=True)
rol['Rol_Id'] = rol.index + 1
order = ['Rol_Id', 'Rol']
rol = rol[order]
rol
```

```python
victim = df[["Víctima"]].drop_duplicates().reset_index(drop=True)
victim['Víctima_Id'] = victim.index + 1
victim.loc[victim['Víctima'] == 'SD', 'Víctima_Id'] = 0
victim = victim.sort_values(by='Víctima_Id').reset_index(drop=True)
order = ['Víctima_Id', 'Víctima']
victim = victim[order]
victim
```

```python
accused = df[["Acusado"]].drop_duplicates().reset_index(drop=True)
accused['Acusado_Id'] = accused.index +1
accused.loc[accused['Acusado'] == 'SD', 'Acusado_Id'] = 0
accused = accused.sort_values(by='Acusado_Id').reset_index(drop=True)
order = ['Acusado_Id', 'Acusado']
accused = accused[order]

accused
```

```python
sex = df[["Sexo"]].drop_duplicates().reset_index(drop=True)
sex['Sexo_Id'] = sex.index + 1
order = ['Sexo_Id', 'Sexo']
sex = sex[order]
sex
```

```python
tipo_calle = df[["Tipo De Calle"]].drop_duplicates().reset_index(drop=True)
tipo_calle['Tipo_Calle_Id'] = tipo_calle.index + 1
order = ['Tipo_Calle_Id', 'Tipo De Calle']
tipo_calle = tipo_calle[order]
tipo_calle
```

```python
crossroad = df[["Cruce"]].drop_duplicates().reset_index(drop=True)
crossroad['Cruce_Id'] = crossroad.index + 1
order = ['Cruce_Id', 'Cruce']
crossroad = crossroad[order]
crossroad
```

```python
time_date = df[['Hechos_Id', 'Año', 'Mes', 'Dia', 'Día Semana', 'Franja Horaria', 'Estación']]

time_date = time_date.merge(season, on='Estación', how='left')
time_date.drop('Estación', axis=1, inplace=True)
time_date.head(3)
```

```python
location = df[['Hechos_Id', 'Pos X', 'Pos Y', 'Calle', 'Barrio']]

location = location.merge(nbh, on='Barrio', how='left')
location.drop('Barrio', axis=1, inplace=True)
location.head(3)
```

```python
accidents_main = df[['Hechos_Id', 'Número De Victimas', 'Fecha', 'Hora', 'Lugar Del Hecho', 'Cruce', 'Tipo De Calle', 'Comuna', 'Participantes', 'Víctima', 'Acusado', 'Rol', 'Sexo', 'Rango Etario', 'Edad']]
accidents_main = accidents_main.merge(commune, on='Comuna', how='left')
accidents_main = accidents_main.merge(participants, on='Participantes', how='left')
accidents_main = accidents_main.merge(victim, on='Víctima', how='left')
accidents_main = accidents_main.merge(accused, on='Acusado', how='left')
accidents_main = accidents_main.merge(rol, on='Rol', how='left')
accidents_main = accidents_main.merge(sex, on='Sexo', how='left')
accidents_main = accidents_main.merge(age_range, on='Rango Etario', how='left')
accidents_main = accidents_main.merge(crossroad, on='Cruce', how='left')
accidents_main = accidents_main.merge(tipo_calle, on='Tipo De Calle', how='left')
columns_to_drop = ['Comuna', 'Tipo De Calle', 'Participantes', 'Rango Etario']
accidents_main.drop(columns_to_drop, axis=1, inplace=True)
order = ['Hechos_Id', 'Número De Victimas', 'Fecha', 'Hora', 'Lugar Del Hecho', 'Cruce_Id',  'Tipo_Calle_Id', 'Comuna_Id', 'Participantes_Id', 'Víctima_Id', 'Acusado_Id', 'Rol_Id', 'Sexo_Id', 'Rango_Etario_Id', 'Edad']
accidents_main = accidents_main[order]
accidents_main
```

CARGAMOS TODOS NUESTROS ARCHIVOS PARTIENDO DE LAS TABLAS CREADAS.

```python
dfs = [accidents_main, participants, location, time_date, crossroad, tipo_calle, sex, accused, victim, rol, age_range, season, nbh, commune ]
# Nombres correspondientes para cada archivo
names = ['hechos_main', 'participantes', 'ubicacion', 'fecha_hora', 'cruce', 'tipo_calle', 'sexo', 'acusado', 'victima', 'rol', 'rango_etario', 'estacion', 'barrios', 'comunas']

save_to_csv(dfs, names)
```
