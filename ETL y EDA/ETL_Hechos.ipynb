Ahora agregamos una nueva columna 'Día Semana'

```python
locale.setlocale(locale.LC_TIME, 'es_ES')

# Añadimos la nueva columna 'Día semana'
hechos['Día Semana'] = hechos['Fecha'].dt.strftime('%A').str.capitalize()

hechos['Mes'] = hechos['Mes'].apply(lambda x: calendar.month_name[x]).str.capitalize()

# Restaurar a la configuración predeterminada
locale.setlocale(locale.LC_TIME, 'C')
# Visualizamos
hechos.head()
```

```python
def get_season(mes):
    if mes in [12, 1, 2]:
        return 'Verano'
    elif mes in [3, 4, 5]:
        return 'Otoño'
    elif mes in [6, 7, 8]:
        return 'Invierno'
    elif mes in [9, 10, 11]:
        return 'Primavera'
    else:
        return 'Desconocido'

# Agregamos nueva columna 'Estación'
hechos['Estación'] = hechos['Fecha'].dt.month.apply(get_season)
hechos
```

Ahora vamos a agregar la palabra 'Comuna' a los valores en la columna 'Comuna'

```python
hechos["Comuna"] = hechos.apply(lambda x: "Comuna "+str(x["Comuna"]), axis=1)
```
